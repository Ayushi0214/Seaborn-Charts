```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset("iris")
print(data)

sns.pairplot(data = data)
```

```python
data = pd.read_excel("ESD.xlsx")

#print(data)
sns.scatterplot(data = data, x = "Age",y = "Annual Salary", hue = "Department")
plt.legend(bbox_to_anchor = (0.2,0, 1.2,1)) #x , y , width, height
plt.show()
```

```python
df = pd.read_excel("ESD.xlsx")

sns.violinplot(data = df, x = "Gender", y = "Annual Salary", orient = "vertical")
plt.show()
```
