```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset("tips")
print(data)

gp = data.groupby("day").agg({"tip":"median"})
sns.heatmap(gp)
plt.show()
```

```python
data = pd.read_excel("food_data.xlsx")
#print(data)
gp = data.groupby("Category").agg({"Protein":"mean", "Fat":"mean","Carbs":"mean", "Fibre":"mean"})
gp = gp.dropna()
print(gp)
sns.heatmap(gp)
plt.show()
```

```python
corr = gp.corr()
print(corr)

sns.heatmap(corr, vmin = -1, vmax = 1, annot = True)
```
